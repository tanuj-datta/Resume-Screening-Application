```python
import re

def cleanResume(txt):
    cleanTxt = re.sub('http\S+\s', '', txt)
    cleanTxt = re.sub('@\S+', ' ', cleanTxt)
    cleanTxt = re.sub('#\S+', ' ', cleanTxt)
    cleanTxt = re.sub('RT|cc', ' ', cleanTxt)
    cleanTxt = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_{|}~ """), ' ', cleanTxt)
    cleanTxt = re.sub(r'[^\x00-\x7f]', ' ', cleanTxt)
    cleanTxt = re.sub('\s+', ' ', cleanTxt)
    return cleanTxt

cleanResume("my website $$$$---++++++++&&&###datta is http://youtube access it @gmail.com")
```

```python
df['Resume'] = df['Resume'].apply(lambda X: cleanResume(X))
df['Resume'][0]
```

```python
#this steps makes the dataset into single vector entities
from sklearn.feature_extraction.text import  TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf.fit(df['Resume'])
requiredText = tfidf.transform(df['Resume'])
```

```python
df
```

```python
 requiredText
```

```python
from sklearn.model_selection import train_test_split
```

```python
#20% data for testing
X_train, X_test, y_train, y_test = train_test_split(requiredText ,df['Category'],test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Here we use KNN Classifier coz it will find the nearest points
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
from IPython.display import display

# Assuming you have X_train, X_test, y_train, and y_test defined

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
ypred = clf.predict(X_test)

# Print the predictions and accuracy score
print("Predictions:")
display(ypred)

print("Accuracy Score:")
accuracy = accuracy_score(y_test, ypred)
print(accuracy)
```

```python
#vector is pickled here
import pickle
pickle.dump(tfidf, open('tfidf.pkl', 'wb'))
pickle.dump(clf, open('clf.pkl', 'wb'))
```
